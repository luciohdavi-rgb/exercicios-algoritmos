# Aula 07: Listas

Até agora, nossas variáveis (`idade`, `nome`, `preco`) guardavam **um único valor** por vez.
Mas e se precisássemos guardar a nota de 50 alunos? Criar 50 variáveis (`nota1`, `nota2`, ..., `nota50`) seria terrível.
Para guardar uma coleção de valores em uma única variável, usamos as **Listas**.

## 1. Criando Listas

Listas são criadas usando colchetes `[]` e separando os itens por vírgulas.
Podem conter números, strings, misturas, e até outras listas!

```python
notas = [8.5, 7.0, 9.2, 5.5, 10.0]
frutas = ["Maçã", "Banana", "Laranja"]
lista_mista = ["João", 25, True, 1.75]

print("Notas:", notas)
print("Frutas:", frutas)
```

## 2. Acessando Itens da Lista

Cada item na lista tem um "endereço" (posição/índice). O Python **começa a contar do 0**.
Isso significa que o 1º item está no índice 0, o 2º no índice 1, etc.

```python
print("Primeira fruta:", frutas[0])
print("Segunda fruta:", frutas[1])

# O Python também permite o uso de índices negativos para contar de trás pra frente!
print("Última fruta:", frutas[-1])
```

## 3. Alterando e Adicionando Itens

Listas em Python são mutáveis, ou seja, podemos trocar, adicionar e remover itens delas quando quisermos.

```python
cesta = ["Pera", "Uva", "Maçã"]
print("Cesta original:", cesta)

# Alterando um item pelo índice
cesta[1] = "Morango"
print("Troquei Uva por Morango:", cesta)

# Adicionando um novo item NO FINAL da lista com .append()
cesta.append("Goiaba")
print("Adicionei Goiaba no final:", cesta)

# Removendo um item da lista com .remove()
cesta.remove("Pera")
print("Removi a Pera:", cesta)
```

## 4. O Poderoso Casamento: `for` + Listas

Listas foram feitas para serem usadas com o laço `for`. Veja como é fácil passar por todos os itens de uma lista:

```python
carrinho = ["Pão", "Leite", "Manteiga", "Café"]

print("Itens do meu carrinho:")
for item in carrinho:
    print(f"- Eu preciso comprar: {item}")
```

### Funções Úteis com Listas (Tamanho, Soma, Máximo)
O Python tem funções embutidas que nos ajudam a manipular dados em listas muito rapidamente.

```python
precos = [10.50, 4.20, 15.00, 2.50]

# len() para ver quantos itens existem na lista
print(f"Quantos itens? {len(precos)}")

# sum() para somar os números
# max() e min() para o maior e menor número
print(f"Total da compra: R$ {sum(precos)}")
print(f"Produto mais caro: R$ {max(precos)}")
print(f"Produto mais barato: R$ {min(precos)}")
```

```python
idades = [25, 45, 18, 30, 22, 60, 52]

# Seu código aqui

media = sum(idades) / len(idades)

print("A média das idades é:", media)

for i in idades:
  if i > media:
    print(i)
```

### Exercício 2: Somar Elementos

### Exercício 3: Encontrar o Maior Número

```python
# Seu código aqui

numeros = [32, 21, 53, 20, 16, 3, 7]
primeiro_numero = numeros[0]
maior = primeiro_numero
menor = primeiro_numero

for i in range (len(numeros)):
  if numeros[i] > maior:
    maior = numeros[i]
  if numeros [i] < menor:
    menor = numeros[i]

print(maior)
print(menor)
```
