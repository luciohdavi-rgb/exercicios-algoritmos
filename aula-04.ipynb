## Como resolver um problema de programação?

1.  **Entenda o problema:** Leia com atenção o que está sendo pedido.
2.  **Identifique as Entradas e Saídas:** O que eu recebo do usuário? O que eu devo mostrar na tela?
3.  **Esboce o Passo a Passo (Algoritmo):** "Primeiro, peço o valor X. Depois verifico se X é maior que 10. Se sim, faço isso."
4.  **Codifique:** Transforme seu passo a passo em código Python.
5.  **Teste:** Execute com os valores de exemplo e veja se funciona.

## Exemplo Resolvido 1: Ano Bissexto

**O problema:** Crie um programa que peça um ano (ex: 2024) e informe se ele é Bissexto ou não.

*Regra do Ano Bissexto:*
Ele precisa ser múltiplo de 4 (`ano % 4 == 0`).
Mas se for fim de século (múltiplo de 100), ele **não** é bissexto, a menos que ele também seja múltiplo de 400.

```python
# Solução do Professor
ano = int(input("Digite o ano: "))

if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
    print(f"O ano {ano} É bissexto!")
else:
    print(f"O ano {ano} NÃO é bissexto.")
```

## Exemplo Resolvido 2: Calculadora Simples

**O problema:** Peça dois números e um operador matemático (`+`, `-`, `*` ou `/`) ao usuário. Imprima o resultado da operação.

```python
num1 = float(input("Digite o primeiro número: "))
op = input("Digite a operação (+, -, *, /): ")
num2 = float(input("Digite o segundo número: "))

if op == "+":
    print(f"Resultado: {num1 + num2}")
elif op == "-":
    print(f"Resultado: {num1 - num2}")
elif op == "*":
    print(f"Resultado: {num1 * num2:.2f}") # Arredondando para duas casas decimais aqui
elif op == "/":
    # Cuidados extras em programação: não se divide por zero!
    if num2 == 0:
        print("Erro: Divisão por zero.")
    else:
        print(f"Resultado: {num1 / num2}")
else:
    print("Operador inválido!")
```

### Desafio 3: Triângulos
Peça os 3 lados de um triângulo. O programa deverá informar se os valores formam um triângulo. Caso formem, diga o tipo do triângulo:
- Equilátero: 3 lados iguais;
- Isósceles: 2 lados iguais;
- Escaleno: 3 lados diferentes.
*Dica: Para formar um triângulo, a soma de quaisquer dois lados precisa ser maior que o terceiro lado.*

```python
# Seu código aqui
lado_a = int(input("Digite o lado A: "))
lado_b = int(input("Digite o lado B: "))
lado_c = int(input("Digite o lado C: "))

if lado_a + lado_b > lado_c and lado_a + lado_c > lado_b and lado_b + lado_c > lado_a: # Condição para formar triângulo
  print("Seus lados formam um triângulo")
  if lado_a == lado_b and lado_b == lado_c: # Se os 3 lados forem iguais = equilátero
    print("Triângulo equilátero")
  elif lado_a != lado_b and lado_b != lado_c and lado_a != lado_c: # Se os 3 forem diferentes = escaleno
    print("Triângulo escaleno")
  else:
    print("Triângulo isósceles")
else:
  print("Seus lados não formam um triângulo")
```

### Exercício 8: Calculadora de IMC com Classificação
Peça o peso (em kg) e a altura (em metros) do usuário. Calcule o Índice de Massa Corporal (IMC) usando a fórmula `IMC = peso / (altura ** 2)`. Em seguida, classifique o IMC de acordo com as seguintes categorias:
- Abaixo do peso: IMC < 18.5
- Peso normal: 18.5 <= IMC < 24.9
- Sobrepeso: 25.0 <= IMC < 29.9
- Obesidade Grau I: 30.0 <= IMC < 34.9
- Obesidade Grau II: 35.0 <= IMC < 39.9
- Obesidade Grau III (Mórbida): IMC >= 40.0

```python
print ("------- Calculadora IMC -------")
peso = float(input("Digite seu peso em Kg: "))
altura = float(input("digite suaalturaaluta em metros: "))

imc = peso / (altura ** 2)
print ("Seu IMC é:", round(imc, 2))

if imc < 18.5:
  print("Abaixo do peso")
elif 18.5 <= imc < 24.9:
  print("Peso normal")
elif 25.0 <= imc < 29.9:
  print("Sobrepeso")
elif 30.0 <= imc < 34.9:
  print("Obesidade grau I")
elif 35.0 <= imc < 39.9:
  print("Obesidade grau II")
elif imc >= 40:
  print("Obesidade mórbida")
```

### Exercício 9: Condição de Empréstimo Bancário
Um banco oferece empréstimos com as seguintes condições:
- Salário mínimo de R$ 2000.
- Idade entre 25 e 60 anos (inclusive).
- Não ter histórico de dívidas (representado por uma variável booleana `tem_dividas = False`).
Peça ao usuário o salário, a idade e se ele possui dívidas (True/False). Verifique se ele se qualifica para o empréstimo usando uma combinação de operadores lógicos.

```python
salario = float(input("Digite seu salário: R$"))
idade = int(input("Digite sua idade: "))
dividas = input("Você possui dívidas? (sim/não): ").lower()

tem_dividas = (dividas == 'sim')

if salario >= 2000 and 25 <= idade <= 60 and not tem_dividas:
    print("Você se qualifica para o empréstimo.")
else:
    print("Você não se qualifica para o empréstimo no momento.")
```
